### The measurement update step:

Calculating the measurement: calculating the new mean and the new variance from two Gaussian 

```python
def update(mean1, var1, mean2, var2):
    new_mean = (var2 * mean1 + var1 * mean2) / (var1 + var2)
    new_var = 1 / (1 / var1 + 1 / var2)
    return [new_mean, new_var]
```

```python
update(10., 4., 12., 4.)
```

### The Motion(Prediction) step:
Calculating the Pridiction(motion):calculating the new mean and the new variance from two Gaussian 

```python
def predict(mean1, var1, mean2, var2):
    new_mean = mean1 + mean2
    new_var = var1 + var2
    return [new_mean, new_var]
```

```python
predict(10., 4., 12., 4.)
```

Implementation of both steps:

```python
def update(mean1, var1, mean2, var2):
    new_mean = (var2 * mean1 + var1 * mean2) / (var1 + var2)
    new_var = 1 / (1 / var1 + 1 / var2)
    return [new_mean, new_var]
def predict(mean1, var1, mean2, var2):
    new_mean = mean1 + mean2
    new_var = var1 + var2
    return [new_mean, new_var]

update(10., 4., 12., 4.)
predict(10., 4., 12., 4.)
```

A main program that takes these two functions, update and predict, and feeds it into a sequence of measurement and motions

```python
measurements = [5., 6., 7., 9., 10.]
motion = [1., 1., 2., 1., 1.]
measurement_sig = 4.
motion_sig = 2.
mu = 0.
sig = 1000.
```

```python
for n in range(len(measurements)):
    [mu, sig] = update(mu, sig, measurements[n], measurement_sig)
    [mu, sig] = predict(mu, sig, motion[n], motion_sig)
```
